## Anonymous web data from www.microsoft.com

* This dataset records which areas (Vroots) of www.microsoft.com each user visited in a one-week timeframe in Feburary 1998.
* Source: Jack S. Breese, David Heckerman, Carl M. Kadie.
* The data was created by sampling and processing the www.microsoft.com logs. The size of the sample was 38000 users randomly-selected.
  * Training Instances	32711
  * Testing Instances	5000
  * Attributes	294
  * nonvisits are implicit (missing). 38000 - (32711 + 5000) = 289
* More information: http://kdd.ics.uci.edu/databases/msweb/msweb.data.html

### Data Format

* Attribute lines:
  * For example, 'A,1277,1,"NetShow for PowerPoint","/stream"'
    
    Where:
    * 'A' marks this as an attribute line, 
    * '1277' is the attribute ID number for an area of the website (called a Vroot),
    * '1' may be ignored, 
    * '"NetShow for PowerPoint"' is the title of the Vroot, 
    * '"/stream"' is the URL relative to "http://www.microsoft.com"

* Case and Vote Lines: For each user, there is a case line followed by zero or more vote lines.
  * For example:
    * C,"10164",10164
    * V,1123,1
    * V,1009,1
    * V,1052,1
      
      Where:
      *   'C' marks this as a case line, 
      *   '10164' is the case ID number of a user, 
      *   'V' marks the vote lines for this case, 
      *   '1123', 1009', 1052' are the attributes ID's of Vroots that a user visited. 
      *   '1' may be ignored

## The goal find the relationship between web pages

```python
import re, requests
import bs4
import pandas as pd
import numpy as np
```

```python
response = requests.get('http://kdd.ics.uci.edu/databases/msweb/msweb.html')
res_bs = bs4.BeautifulSoup(response.content, 'html.parser')
```

```python
saved_link = []
for a_path in res_bs.find_all('a'):
    if a_path.get('href'):
        #print(a_path)
        if re.search(r'.*\.gz', a_path.get('href')):
            #print(a_path.get('href'))
            #print('http://kdd.ics.uci.edu/databases/msweb/'+ i.get('href'))
            link_to_download = 'http://kdd.ics.uci.edu/databases/msweb/'+ a_path.get('href')
            saved_link.append(link_to_download)

for link_to_download in saved_link:
    name_file = link_to_download.split('/')[-1]
    
    response = requests.get(link_to_download)
    
    with open(f'{name_file}', 'wb') as writer:
        writer.write(response.content)
```

```python
import gzip

with gzip.open(r'C:\Users\LENOVO\Desktop\projects\3.- STS\anonymous-msweb.data.gz', 'rb') as reader:
    data = reader.read()
    
data_dec = data.decode()
```

```python
data_a = []
data_c = {}
for i, line in enumerate(data_dec.splitlines()):
    if re.search(r'^A.*', line):
        string_a = re.search('^A.*', line).group()
        list_a = string_a.split(',')
        list_a.remove('A')
        list_a.remove('1')
        data_a.append(list_a)
    elif re.search(r'^C.*', line):
        string_c = re.search(r'^C.*(\d{5}).*', line).group(1)
        data_c[string_c] = i
        
```

```python
data_attribute = pd.DataFrame(data_a, columns=['attribute_ID', 'title', 'URL'])
```

```python
data_attribute.URL = data_attribute.URL.str.replace('"', '')
data_attribute.URL = "http://www.microsoft.com" + data_attribute.URL
```

```python
from collections import namedtuple
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def func_c():
    new_data_c = []
    x_in = '10001'
    y_in = data_c[x_in]
    for x, y in data_c.items():
        
        select_list = data_dec.splitlines()[y_in+1:y]
        tup_walz = namedtuple('walz', ['user_ID', 'attribute_ID'])
        for sl in select_list:
            data_walz = tup_walz(user_ID=x_in, attribute_ID=sl)
            new_data_c.append(data_walz)
        y_in = y
        x_in = x
        
    return new_data_c
```

```python
import time
time_init = time.time()
new_data_c = func_c()
time_out = time.time() - time_init
print(time_out/60)
```

```python
data_user = pd.DataFrame(new_data_c)
data_user.attribute_ID = data_user.attribute_ID.str.split(',').str.get(1)
cross_user_attr = pd.crosstab(data_user['user_ID'], data_user['attribute_ID'])
```

```python
data_user_group = data_user.groupby('user_ID')
```

```python
from itertools import combinations
```

```python
user_ID_uniq = data_user.user_ID.unique()

data_final = []
for user in user_ID_uniq:
    data_attr = data_user_group.get_group(user)['attribute_ID']
    data_attr_ls = data_attr.values.tolist()
    comb_ls = combinations(data_attr_ls, 2)
    data_final.extend(comb_ls)


Pages = pd.DataFrame(data_final, columns=['Pag_1', 'Pag_2'])
data_visit = Pages.groupby('Pag_1')['Pag_2'].value_counts().sort_values(ascending=False) 
data_visit.name = 'visits'
```

```python
from matplotlib import cm 

fig, ax = plt.subplots(figsize = (10, 6))

data_visit_to_plot = data_visit.reset_index().astype(np.int32).sort_values(by = 'visits')
x, y = data_visit_to_plot.Pag_1, data_visit_to_plot.Pag_2
c = data_visit_to_plot.visits.copy()
s = data_visit_to_plot.visits.copy()*1.4

ax.scatter(x, y,s = s,  c = c, cmap = cm.Blues, alpha = 0.5 )
ax.set(xlabel='Page $1$', ylabel='Page $2$')
x_s, y_s = [1008, 1008], [1034, 1009] 
texts = ['(Free Downloads,\n Internet Explorer)', '(Free Downloads,\n Windows Family of OSs)' ]
for x_t, y_t, txt in zip(x_s, y_s, texts): 
    ax.text(x_t, y_t, txt, color = 'white',ha = 'center', va = 'center' , fontsize = 8)
plt.show()
```

The pair of web pages most visited was `Free Downloads and Internet Explores`. The $15\%$ of the visits that did the people selected were made up of this combination of web pages.

Also, we can see the pair web pages that were most visited were around of  attibutte_id 1000 and 1050.
